```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
res_path = 'synthetic_sample_4000'
directory = "results/synthetic_sample_4000"


plt.figure(figsize=(10, 6))
for filename in os.listdir(directory):
    if filename.startswith("incidence_seed_") and filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        data = pd.read_csv(file_path, sep='\t')
        plt.plot(data['H1N1'], label=filename)

plt.xlabel('Days')
plt.ylabel('Incidence')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(10, 6))
for filename in os.listdir(directory):
    if filename.startswith("prevalence_seed_") and filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        data = pd.read_csv(file_path, sep='\t')
        plt.plot(data['H1N1'], label=filename)

plt.xlabel('Days')
plt.ylabel('Prealence')
plt.legend()
plt.grid()
```
